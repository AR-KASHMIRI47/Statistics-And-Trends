```python
import pandas as pd
import matplotlib.pyplot as plt

# With function Also

def plot_graph(file_path):
    
    """ Load the dataset
        check the dataset columns
        plot the graph of session duration columns and calories burned with figure size as well
        plot the scatter graph make a title with x and y lables with adjustable font size.
        
    """ 
    df = pd.read_csv(file_path)
    
    # Check if the required columns are present
    if 'Session_Duration (hours)' in df.columns and 'Calories_Burned' in df.columns:
        # Create the scatter plot
        plt.figure(figsize=(10, 6))
        plt.scatter(df["Session_Duration (hours)"], df["Calories_Burned"], alpha=0.7, color='dodgerblue')
        
        # Add titles and labels for clarity
        plt.title("Calories Burned vs. Session Duration", fontsize=16)
        plt.xlabel("Session Duration (hours)", fontsize=12)
        plt.ylabel("Calories Burned", fontsize=12)
        
        # Show the plot
        plt.grid(True)
        plt.show()
    else:
        print("Error: The dataset does not contain the required 'Session_Duration' and 'Calories_Burned' columns.")

# Example usage
file_path = "C:/Users/Abdul/Downloads/archive/gym_members_exercise_tracking.csv"
plot_graph(file_path)
```
