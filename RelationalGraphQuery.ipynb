```python
file_path = "C:/Users/Abdul/Downloads/archive/gym_members_exercise_tracking.csv"

# with function
def visualize(file_path):
    # upload the dataset
    # filepath is a path of the file contain dataset
    
    df = pd.read_csv(file_path)

    # Plots a bar chart of the average calories burned per workout type from a given dataset
    # check Age and BMI columns 
    # figure size 
    # plot type and x and y lable with font size
    
    if 'Age' in df.columns and 'BMI' in df.columns:
        plt.figure(figsize=(8, 6))
        plt.title("Relationship between Age and BMI", fontsize=16)
        plt.scatter(df['Age'], df['BMI'], alpha=1)
        plt.xlabel("Age", fontsize=14)
        plt.ylabel("BMI", fontsize=14)
        plt.grid(True)
        plt.show()
    else:
        # not found error handling
        print("Error: The dataset does not contain the required 'Age' and 'BMI' columns.")

# Use the function with the file path
visualize(file_path)
```

```python
correlation = df['Age'].corr(df['BMI'])
print(f'Correlation between Age and BMI: {correlation:.2f}')
```
