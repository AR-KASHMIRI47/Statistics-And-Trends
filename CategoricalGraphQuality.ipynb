```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/Abdul/Downloads/archive/gym_members_exercise_tracking.csv")
print(df.columns)
```

```python
average_calories = df.groupby('Workout_Type')['Calories_Burned'].mean()
average_calories
```

```python
average_calories.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Calories Burned per Workout', fontsize=12)
plt.xlabel('Workout Type', fontsize=14)
plt.ylabel('Average Calories Burned', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate for readability
plt.yticks(fontsize=12)
plt.grid(True)

# Display the plot
plt.show()
```

```python
file_path = "C:/Users/Abdul/Downloads/archive/gym_members_exercise_tracking.csv"
#with function also

#using pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

def plot_average_calories(file_path):
    
    # file path is the path of the file contain dataset
    # Load the dataset
    # check if workout type and calories burned columns are available in dataset 
    # make a group of all workout type with calories burned and differentiate the mean of it.
    # plot the figure and x and y lables with font size adjustable as well as rotations of the x-axis label with grid true 
    # Error Handling as well
    
    df = pd.read_csv(file_path)

    if 'Workout_Type' in df.columns and 'Calories_Burned' in df.columns:
        average_calories = df.groupby('Workout_Type')['Calories_Burned'].mean()
        plt.figure(figsize=(8, 6))
        average_calories.plot(kind='bar')
        plt.title('Average Calories Burned per Workout', fontsize=8)
        plt.xlabel('Workout Type', fontsize=10)
        plt.ylabel('Average Calories Burned', fontsize=10)
        plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate for readability
        plt.yticks(fontsize=12)
        plt.grid(True)
        plt.show()
    else:
        print("Error: The dataset does not contain the required 'Workout_Type' and 'Calories_Burned' columns.")

plot_average_calories(file_path)
```
